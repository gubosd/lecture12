- Iris 데이터에서 각 속성의 평균을 뺀 값을 구해보자
- 먼저 속성들의 평균들을 array 로 만들고 뺄셈을 적용한다

```python
iris = np.loadtxt('iris.csv', skiprows=1, delimiter=',', usecols=[0,1,2,3])
```

```python
iris.mean(axis=0)
```

```python
iris  - iris.mean(axis=0) # axis=0 이면 아래로 계산
```

```python
iris_norm = (iris  - iris.mean(axis=0))/iris.std(axis=0) # 정규화
```

```python
iris_norm.mean(axis=0)
```

```python
iris_norm.std(axis=0)
```

```python
plt.plot(iris)
```

```python
plt.figure(figsize=[10,4])

plt.subplot(1,2,1)
plt.plot(iris)
plt.title('original')

plt.subplot(1,2,2)
plt.plot(iris_norm, alpha=0.8)
plt.legend(['sl','sw','pl','pw'])
plt.hlines([0,1,-1],0,150,linestyles=':')
plt.title('normed')
pass
```

### 통계함수
- min(), max()
- sum()
- mean(), std()

$$ 표준편차 = \sqrt{ {1 \over N} \sum_{i=0}^{N-1} (x_i - \bar{x})^2 },\quad \bar{x} = {1 \over N} \sum_{i=0}^{N-1} {x_i} $$

- 표준편차의 제곱이 분산이다.

```python
a = np.array([1,2,3,4,5])

m = a.sum()/len(a) # a.mean()
s = np.sqrt( np.sum( (a-m)**2 ) / len(a)  )

display(m,s)
```

```python
a - m
```

```python
(a-m)**2
```

```python
np.sum((a-m)**2)
```

```python
np.sum((a-m)**2) / len(a)
```

```python
np.sqrt(np.sum((a-m)**2) / len(a))
```

```python
a.std()
```

```python
a2 = (a-m)/s # 정규화 (평균을 빼고 표준편차로 나누어 준다)
a2
```

```python
a2.mean(), a2.std()
```
